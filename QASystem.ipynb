```python
#import libraries
from transformers import pipeline
import wikipediaapi
import re

```

```python
def get_wiki_context(page_name, lang='en'):

    user_agent = "My Application's Name: (Question Answering Application)"
    wiki = wikipediaapi.Wikipedia(language=lang, user_agent=user_agent)

    page = wiki.page(page_name)

    if not page.exists():
        print("The page does not exist.")
        return None

    return page.text
```

```python

def find_sentence_with_answer(context, answer):
    # Split the context into sentences
    sentences = re.split(r'(?<=[.!?])\s+', context)

    # Find and return the sentence that contains the answer
    for sentence in sentences:
        if answer in sentence:
            return sentence
    return answer
```

```python
def question_answering(context):
    # Instantiate a Question Answering pipeline
    nlp = pipeline('question-answering', model='distilbert-base-uncased-distilled-squad')

    # Print the context
    print("*** This is a QA system by Group 7. It will try to answer questions\n"
          "*** from the Wikipedia page of your choice. \n"
          "*** You can enter 'exit' to leave the program. \n")
    while True:
        # Get the question from the user
        question = input("Please enter your question: ")

        # Check if the user wants to quit
        if question.lower() == "exit":
            print("Thank you! Goodbye!")
            break

        # Prepare the inputs
        inputs = {
            'question': question,
            'context': context
        }

        # Get the answer from the model
        answer = nlp(inputs)

        # Find the sentence that contains the answer
        sentence_with_answer = find_sentence_with_answer(context, answer['answer'])

        print(f"Answer: {sentence_with_answer}\n")
```

```python
# Test the function
page_name = 'New_York_City'  # Use the Wikipedia page name
context = get_wiki_context(page_name)

if context:
    question_answering(context)
```
